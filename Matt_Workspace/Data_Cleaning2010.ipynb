```python
import pandas as pd
```

```python
months = ['02', '03', '04', '05', '06']
```

```python
df = pd.read_parquet('Yellow2015/yellow_tripdata_2015-01.parquet', engine='pyarrow')

df['date'] = df['tpep_pickup_datetime'].dt.date

df['Trip_Length'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']

dfgrouped = df.groupby('date').agg({'passenger_count':['sum', 'count'], 'trip_distance':['sum', 'mean'], 'total_amount':'sum', 'Trip_Length':['sum', 'mean']})
```

```python
for month in months:
    df = pd.read_parquet('Yellow2015/yellow_tripdata_2015-'+month+'.parquet')
    df['date'] = df['tpep_pickup_datetime'].dt.date
    df['Trip_Length'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
    tempdf = df.groupby('date').agg({'passenger_count':['sum', 'count'], 'trip_distance':['sum', 'mean'], 'total_amount':'sum', 'Trip_Length':['sum', 'mean']})
    dfgrouped = pd.concat([dfgrouped, tempdf], axis = 0)
```
