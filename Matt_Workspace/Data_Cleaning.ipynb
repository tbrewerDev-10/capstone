```python
df[['Date', 'Pickup_Time']] = df.Trip_Pickup_DateTime.str.split(' ', expand=True)
```

```python
df.drop('Pickup_Time', axis = 1, inplace=True)
df.info()
```

```python
df.Trip_Pickup_DateTime = pd.to_datetime(df.Trip_Pickup_DateTime)
df.Trip_Dropoff_DateTime = pd.to_datetime(df.Trip_Dropoff_DateTime)
```

```python
df['Trip_Length'] = df['Trip_Dropoff_DateTime'] - df['Trip_Pickup_DateTime']
```

```python
df['Payment_Type'].unique()
```

```python
df1 = df.groupby('Date').agg({'Passenger_Count':['sum', 'count'], 'Trip_Distance':'sum', 'Total_Amt':'sum', 'Trip_Length':'sum'})
```

```python

from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import linear_model
import statistics
```

```python
df1.reset_index(inplace=True)
y = df1[('Total_Amt', 'sum')]
X = df1.drop([('Total_Amt', 'sum'), ('Date', )], axis = 1)
```

```python
X[('Trip_Length', 'sum')]= X[('Trip_Length', 'sum')].astype('timedelta64[m]')
X.info()
```

```python
model = linear_model.LinearRegression()

x = 0
temp = []
for i in range(100):
    Xt, Xv, yt, yv = train_test_split(X, y, test_size=.2, random_state=x)
    model.fit(Xt, yt)
    temp.append(model.score(Xv, yv))
    x += 1

print(statistics.mean(temp))
print(max(temp))
print(min(temp))
# plot_tree(model)
```
